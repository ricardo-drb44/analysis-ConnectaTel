## 🧩Paso 3: Limpieza básica de datos

### 3.1 Corregir sentinels y fechas imposibles
**🎯 Objetivo:**  
Aplicar reglas de limpieza para reemplazar valores sentinels y corregir fechas imposibles.

**Instrucciones:**  
- En `age`, reemplaza el sentinel **-999** con la mediana.
- En `city`, reemplaza el sentinel `"?"` por valores nulos (`pd.NA`).  
- Marca como nulas (`pd.NA`) las fechas fuera de rango.

```python


# Reemplazar -999 por la mediana de age
age_mediana = users.loc[users['age'] != -999, 'age'].median()
users['age'] = users['age'].replace(-999, age_mediana)

# Verificar cambios
users['age'].describe()

```

```python


# Reemplazar ? por NA en city
users['city'] = users['city'].replace('?', pd.NA)
# Verificar cambios
print("¿Quedan caracteres '?' en city?:", (users['city'] == '?').sum())
print("Cantidad de nulos en city:", users['city'].isna().sum())

```

```python


# Marcar fechas futuras como NA para reg_date
users.loc[users['reg_date'].dt.year > 2024, 'reg_date'] = pd.NaT
# Verificar cambios
print(users['reg_date'].dt.year.value_counts(dropna=False))

```

### 3.2 Corregir sentinels y fechas imposibles
**🎯 Objetivo:**  
Decidir qué hacer con los valores nulos según su proporción y relevancia.

**Instrucciones:**
- Verifica si los nulos en `duration` y `length` son **MAR**(Missing At Random) revisando si dependen de la columna `type`.  
  Si confirmas que son MAR, **déjalos como nulos** y justifica la decisión.

```python


# Verificación MAR en usage (Missing At Random) para duration
print("Distribución de nulos en 'duration' según el tipo de registro:")
print(usage.groupby('type')['duration'].apply(lambda x: x.isna().sum()))

print("\nProporción de nulos en 'duration' según el tipo de registro:")
print(usage.groupby('type')['duration'].apply(lambda x: x.isna().mean()))

```

```python


# Verificación MAR en usage (Missing At Random) para length
print("Distribución de nulos en 'length' según el tipo de registro:")
print(usage.groupby('type')['length'].apply(lambda x: x.isna().sum()))

print("\nProporción de nulos en 'length' según el tipo de registro:")
print(usage.groupby('type')['length'].apply(lambda x: x.isna().mean()))

```

Haz doble clic aquíy escribe que tu diagnostico de nulos en `duration` y `length`
Los resultados confirman con precisión matemática que estamos ante nulos de tipo MAR (Missing At Random), los cuales responden a una lógica estructural del negocio y no a pérdidas aleatorias de información:

Columna duration: Tiene un ~100% de nulos cuando el tipo de registro es text. Esto es correcto, ya que los mensajes de texto no poseen una duración en minutos.

Columna length: Tiene un ~100% de nulos cuando el tipo de registro es call. Esto también es correcto, dado que las llamadas telefónicas no tienen una longitud de caracteres.

Acción Recomendada: No se debería eliminar estos nulos ni borrarlos con .dropna().

```python



# Columnas auxiliares
usage["is_text"] = (usage["type"] == "text").astype(int) #conocer el total de mensajes
usage["is_call"] = (usage["type"] == "call").astype(int) #conocer el total de llamadas


# Agrupar información por usuario
usage_agg = usage.groupby('user_id').agg({
    'is_text': 'sum',
    'is_call': 'sum',
    'duration': 'sum',
    'length': 'sum'
}).reset_index()

# observar resultado
usage_agg.head(3)


```

```python


# Renombrar columnas
usage_agg.columns = ['user_id', 'total_texts', 'total_calls', 'total_duration', 'total_length']
# observar resultado
usage_agg.head(3)

```

```python




# Combinar la tabla agregada con el dataset de usuarios
user_profile = pd.merge(users, usage_agg, on='user_id', how='left')
user_profile.head(5)



```

-Age: No presenta outliers. Los datos de edad están distribuidos de manera uniforme y limpia en ambos planes.
-cant_mensajes: Sí presenta outliers en el extremo superior. El plan Premium muestra mayor dispersión y valores atípicos más altos que el plan Básico.
-cant_llamadas: Sí presenta outliers en el extremo superior. Se observan usuarios con un volumen de llamadas excepcionalmente alto, principalmente en el plan Premium.
-cant_minutos_llamada: Sí presenta outliers en el extremo superior. Existen usuarios de alto consumo ("heavy users") que exceden por mucho el comportamiento del promedio en ambos

```python



# Calcular límites con el método IQR
columnas_limites = ['total_texts', 'total_calls', 'total_duration']

for col in columnas_limites:
    Q1 = user_profile[col].quantile(0.25)
    Q3 = user_profile[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Calculamos solo el límite superior ya que los outliers son positivos/altos
    limite_superior = Q3 + (1.5 * IQR)
    
    print(f"Variable: {col}")
    print(f"  Límite Superior: {limite_superior:.2f}")
    
    # Contamos cuántos usuarios superan este límite
    cantidad_outliers = user_profile[user_profile[col] > limite_superior].shape[0]
    print(f"  Cantidad de registros detectados como outliers: {cantidad_outliers}\n")


```

```python
# Revisa los limites superiores y el max, para tomar la decisión de mantener los outliers o no
user_profile[columnas_limites].describe()
```

💡Insights: 
- cant_mensajes: Mantener. Los valores máximos representan a usuarios reales con un comportamiento intensivo de envío de mensajes ("heavy users"). Eliminarlos sesgaría el análisis del verdadero potencial de consumo en los planes.

-cant_llamadas: Mantener. Son registros válidos que reflejan la actividad de clientes altamente activos. Es información crítica para entender el uso de la red y el comportamiento del segmento Premium.

-cant_minutos_llamada: Mantener. Estos picos de duración reflejan llamadas prolongadas reales y no errores del sistema. Son esenciales para evaluar si los límites actuales de los planes son adecuados o si se necesitan nuevas ofertas comerciales.

### 6.1 Segmentación de Clientes Por Uso

🎯 **Objetivo:** Clasificar a cada usuario en un grupo de uso (Bajo uso, Uso medio, Alto uso) basándose en la cantidad de llamadas y mensajes registrados.

**Instrucciones:**  
- Crea una nueva columna llamada `grupo_uso` en el dataframe `user_profile`.
- Usa comparaciones lógicas (<, >) para evaluar las condiciones de llamadas y mensajes y asigna:
  - `'Bajo uso'` cuando llamadas < 5 y mensajes < 5
  - `'Uso medio'` cuando llamadas < 10 y mensajes < 10
  - `'Alto uso'` para el resto de casos

```python

# Crear columna grupo_uso

import numpy as np

condiciones = [
    (user_profile['total_calls'] < 5) & (user_profile['total_texts'] < 5),
    (user_profile['total_calls'] < 10) & (user_profile['total_texts'] < 10)
]

opciones = ['Bajo uso', 'Uso medio']
user_profile['grupo_uso'] = np.select(condiciones, opciones, default='Alto uso')
```
